```python
model = stGRL(adata,device=device,n_top_genes=4000,dim_output=256)
adata = model.train()
from utils import clustering
tool = 'mclust'
radius = 50
if tool == 'mclust':
   clustering(adata, n_clusters, radius=radius, method=tool, refinement=True)
```

```python
adata = adata[:, adata.var['highly_variable'] == True]
ReX = adata.obsm['rec']
ReX[ReX<0] = 0
adata.layers['stGRL'] = ReX
```

```python
gene_list = ['VAT1L','PCP4','NEFH','CALB1','GNAL','CRYAB']
fig, axs = plt.subplots(1, len(gene_list), figsize=(5 * len(gene_list), 5))

# 循环遍历基因列表并绘制每个基因的图形
for i, plot_gene in enumerate(gene_list):
    sc.pl.spatial(
        adata, 
        img_key="hires", 
        color=plot_gene, 
        cmap='viridis', 
        show=False, 
        ax=axs[i], 
        title='stGRL_'+plot_gene, layer='stGRL', vmax='p99'
    )
    axs[i].collections[-1].colorbar.remove()

# 显示图形
plt.tight_layout()
```
